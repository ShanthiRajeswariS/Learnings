# Ridge plot

Ridge plots are useful for **visualizing the distribution of a quantitative variable across several categories**. They show the density estimates for each category, stacked vertically with an offset, creating a compelling visual comparison of the distributions

```python
# Create a FacetGrid with the 'day' variable on the rows
# Reverse the order to have the first day at the top
g = sns.FacetGrid(df, row="day", hue="day", aspect=15, height=0.5, palette="viridis", row_order=['Thur', 'Fri', 'Sat', 'Sun'])

# Draw the density plots
# Use fill=True to fill the areas under the curves
# alpha controls the transparency
# lw controls the line width
# cut=0 ensures the density plot doesn't extend beyond the data range
g.map(sns.kdeplot, "total_bill", fill=True, alpha=0.8, lw=1.5, cut=0)

# Add a horizontal line at y=0 for each subplot to serve as a baseline
g.map(plt.axhline, y=0, lw=2, clip_on=False)

# Configure the axes and titles
for i, ax in enumerate(g.axes.flat):
    ax.text(-5, 0.05, df['day'].unique()[i],
            fontsize=14, fontweight='bold', color=ax.lines[-1].get_color()) # Add day labels

# Remove axis labels and ticks
g.set_titles("")
g.set_axis_labels("", "Density")
g.set(yticks=[])
g.despine(bottom=True, left=True) # Remove spines

# Adjust the plot to prevent titles overlapping
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
from scipy.stats import gaussian_kde # To calculate KDE
import plotly.figure_factory as ff # Can also be used for distributions
```

```python
# Sort days for consistent order
day_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=day_order, ordered=True)
tips = df.sort_values('day')

fig = go.Figure()

# Define an offset for stacking the plots
offset = 0.15

# Calculate KDE and add a filled trace for each day
for i, day in enumerate(day_order):
    day_data = tips[tips['day'] == day]['total_bill'].dropna()

    if len(day_data) > 1: # Need at least 2 points for KDE
        # Calculate KDE
        kde = gaussian_kde(day_data)
        x_vals = np.linspace(day_data.min() - 1, day_data.max() + 1, 200) # Range for KDE plot
        y_vals = kde(x_vals)

        # Add the filled area trace
        fig.add_trace(go.Scatter(
            x=x_vals,
            y=y_vals + i * offset, # Add offset for stacking
            mode='lines',
            fill='tozeroy', # Fill to the x-axis (which is offset here)
            name=day,
            fillcolor='rgba(0, 100, 80, 0.7)', # Example fill color with transparency
            line=dict(color='rgba(0, 0, 0, 0.5)', width=1), # Example line color
        ))
        # Add a line trace for the density curve outline (optional, for clarity)
        fig.add_trace(go.Scatter(
             x=x_vals,
             y=y_vals + i * offset,
             mode='lines',
             showlegend=False,
             line=dict(color='black', width=1.5)
        ))


# Update layout for a Ridge plot appearance
fig.update_layout(
    title="Total Bill Distribution by Day (Plotly Ridge Plot)",
    xaxis_title="Total Bill",
    yaxis_title="Density", # Y-axis values are offset densities
    showlegend=True,
    hovermode='x unified', # Improve hover behavior
    height=600, # Adjust height as needed
    yaxis=dict(
        showticklabels=False # Hide y-axis ticks
    ),
    xaxis=dict(
        showgrid=False # Hide x-axis grid lines
    ),
    margin=dict(l=0, r=0, t=40, b=0) # Adjust margins
)

fig.show()
```

```python
day_order = ['Thur', 'Fri', 'Sat', 'Sun']
tips['day'] = pd.Categorical(tips['day'], categories=day_order, ordered=True)
tips = tips.sort_values('day')

fig = go.Figure()

# Define a simpler offset
offset = 0.1

# Calculate KDE and add a filled trace for each day
for i, day in enumerate(day_order):
    day_data = tips[tips['day'] == day]['total_bill'].dropna()

    if len(day_data) > 1: # Need at least 2 points for KDE
        kde = gaussian_kde(day_data)
        # Use a fixed number of points for KDE evaluation
        x_vals = np.linspace(day_data.min() - 1, day_data.max() + 1, 100)
        y_vals = kde(x_vals)

        # Add the filled area trace with minimal styling
        fig.add_trace(go.Scatter(
            x=x_vals,
            y=y_vals + i * offset,
            mode='lines',
            fill='tozeroy',
            name=day,
            # Simple color based on index, add some transparency
            fillcolor=f'rgba({i*50}, {i*100}, {255 - i*50}, 0.6)',
            line=dict(width=0), # Hide the line outline for simplicity
        ))


# Update layout for a basic Ridge plot appearance
fig.update_layout(
    title="Total Bill Distribution by Day (Plotly)",
    xaxis_title="Total Bill",
    yaxis_title="Density (Offset)",
    showlegend=True,
    yaxis=dict(showticklabels=False), # Hide y-axis ticks
    height=500,
)

fig.show()
```

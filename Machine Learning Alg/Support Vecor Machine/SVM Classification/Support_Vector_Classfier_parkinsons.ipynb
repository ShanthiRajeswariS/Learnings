```python
#droping a particular column, axis =1
dataset = dataset.drop(['name'], axis=1)
dataset = pd.get_dummies(dataset, prefix_sep='sex')
dataset.head()
```

```python
#finding correlation between the features
corr_var=dataset.corr()
print(corr_var)
```

```python
plt.figure(figsize=(20,17.5))
sns.heatmap(corr_var, annot=True, cmap='BuPu')
```

```python
X = dataset.loc[:, dataset.columns !="status"]
y = dataset["status"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state =0)
print ("Dimention for X train:", X_train.shape)
print ("Dimention for X test:", X_test.shape)
print ("Dimention for y train:", y_train.shape)
print ("Dimention for y test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler().fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn import svm
cl = svm.SVC(kernel='linear', C=0.01)
cl.fit(X_train, y_train)

```

```python
y_pred = cl.predict(X_train)
```

```python
y_pred_1 = cl.predict(X_test)
y_pred_1
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix
cm = confusion_matrix(y_test, y_pred_1)
cm
```

```python
acc = accuracy_score(y_test, y_pred_1)
acc
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C': [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100, 500, 1000, 5000],
              'degree': [2, 3, 4, 5],
              'gamma':[0.001, 0.01, 0.1, 0.5, 1],
              'kernel': ['rbf','poly']
              }
cl = svm.SVC()
grid = GridSearchCV(cl, parameters, cv=10)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
from sklearn.metrics import classification_report
grid_prediction = grid.predict(X_test)
print(classification_report(y_test, grid_prediction))
```

```python
# Evaluate the best model
best_svr = grid.best_estimator_
y_test_pred = best_svr.predict(X_test)
best_test_conf_matrix = confusion_matrix(y_test, y_test_pred)
best_test_acc= accuracy_score(y_test, y_test_pred)
```

```python
print(f"\nBest Model Performance:")
print(f"Test confustion Matrix: {best_test_conf_matrix}")
print(f"Test Accuracy: {best_test_acc:.4f}")
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.inspection import permutation_importance
```

```python
feature_importance = permutation_importance(best_svr, X_test, y_test,
                               n_repeats=10, random_state=42, n_jobs=-1)
```

```python
perm_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance.importances_mean,
    'Std': feature_importance.importances_std
})
```

```python
perm_importance_df = perm_importance_df.sort_values('Importance', ascending=False)
print(perm_importance_df.head(10))

```

```python
# Visualize the top 10 features
plt.figure(figsize=(10, 6))
plt.barh(perm_importance_df['Feature'][:10], perm_importance_df['Importance'][:10])
plt.xlabel('Permutation Importance')
plt.title('Top 10 Features (Permutation Importance)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('permutation_importance.png')
```

```python
# Select two features for visualization based on Feature Importance
X_2d = X[['MDVP:Fo(Hz)', 'MDVP:Flo(Hz)']]
sc_2d = StandardScaler().fit(X_2d)
X_2d_transformed = sc_2d.transform(X_2d)

reduced_svr = svm.SVC(**best_svr.get_params())  # Use the parameters of the best_svr
reduced_svr.fit(sc_2d.transform(X_2d), y)

DecisionBoundaryDisplay.from_estimator(
        reduced_svr,
        X_2d_transformed,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel='MDVP:Fo(Hz)',
        ylabel='MDVP:Flo(Hz)',
    )

# Scatter plot
plt.scatter(X_2d['MDVP:Fo(Hz)'], X_2d['MDVP:Flo(Hz)'],
            c=y,
            s=20, edgecolors="k")
plt.show()
```
